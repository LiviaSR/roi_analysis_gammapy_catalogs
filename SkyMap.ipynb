```python
import pandas as pd
import matplotlib.pyplot as plt

from astropy.coordinates import SkyCoord, Angle, Galactic
from astropy import units as u
```

```python
systems = pd.read_csv('SpiderSystems.csv', delimiter=';', header=0)
```

```python
black_widow_df = systems[systems["Type"] == "BW"]
redback_df = systems[systems["Type"] == "RB"]
```

# Equatorial coordinates

Extract RA and DEC from the CSV file. 

0 < RA < 360 degrees 

-90 < DEC < 90 degrees

```python
#ra = systems['RA (deg)']
#dec = systems['DEC (deg)']

ra_bw = black_widow_df['RA (deg)']
dec_bw = black_widow_df['DEC (deg)']

ra_rb = redback_df['RA (deg)']
dec_rb = redback_df['DEC (deg)']
```

Multiply RA and DEC values by degrees to get an astropy "Quantity" with units of degrees 

```python
#coord = SkyCoord(ra=ra, dec=dec, unit='deg', frame='icrs')

coord_bw = SkyCoord(ra=ra_bw, dec=dec_bw, unit='deg', frame='icrs')
coord_rb = SkyCoord(ra=ra_rb, dec=dec_rb, unit='deg', frame='icrs')
```

Matplotlib needs the coordinates in radians:

```python
#ra_rad = coord.ra.wrap_at(180 * u.deg).radian
#dec_rad = coord.dec.radian


ra_bw_rad = coord_bw.ra.wrap_at(180 * u.deg).radian
dec_bw_rad = coord_bw.dec.radian

ra_rb_rad = coord_rb.ra.wrap_at(180 * u.deg).radian
dec_rb_rad = coord_rb.dec.radian
```

```python
plt.figure(figsize=(8,4.2))
plt.subplot(projection="aitoff")
plt.plot(ra_bw_rad, dec_bw_rad, 'o', color='black', markersize=5, alpha=0.5, label="Black Widow")
plt.plot(ra_rb_rad, dec_rb_rad, 'o', color='red', markersize=5, alpha=0.8, label="Redback")
plt.subplots_adjust(top=0.95,bottom=0.0)
plt.title("Spider systems")
plt.xlabel("Right ascencion (deg)")
plt.ylabel("Declination (deg)")
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1))
plt.grid(True)
plt.savefig('SkyMap_Equatorial.svg')
plt.show()
```







# Galactic coordinates

```python
l_bw = black_widow_df['l (deg)']
b_bw = black_widow_df['b (deg)']

l_rb = redback_df['l (deg)']
b_rb = redback_df['b (deg)']
```

```python
galactic_bw = SkyCoord(l=l_bw, b=b_bw, unit='deg', frame=Galactic)
```

```python
l_bw_rad = galactic_bw.l.wrap_at(180 * u.deg).radian
b_bw_rad = galactic_bw.b.radian
```

```python
galactic_rb = SkyCoord(l=l_rb, b=b_rb, unit='deg', frame=Galactic)
```

```python
l_rb_rad = galactic_rb.l.wrap_at(180 * u.deg).radian
b_rb_rad = galactic_rb.b.radian
```

```python
plt.figure(figsize=(8,4.2))
plt.subplot(projection="aitoff")
plt.plot(l_bw_rad, b_bw_rad, 'o', color='black', markersize=5, alpha=0.5, label="Black Widow")
plt.plot(l_rb_rad, b_rb_rad, 's', color='red', markersize=5, alpha=0.8, label="Redback")
plt.subplots_adjust(top=0.95,bottom=0.0)
plt.title("Spider systems")
plt.xlabel("Galactic longitude (deg)")
plt.ylabel("Galactic latitude (deg)")
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1))
plt.grid(True)
plt.savefig('SkyMap_Galactic.svg')
plt.show()
```
